```python
from pandas import json_normalize 
import requests

def get_data(currencyPair, startDate):
    url = 'https://api.binance.com/api/v3/klines' #the root URL

    
    params = {
      'symbol': currencyPair,
      'startTime': startDate,
      'interval': '1h',  
      'limit': 75
    }

    response = requests.get(url, params=params)
    print(response)
    print(response.json())
    return response.json()
```
